#1. You have a pandas DataFrame df with columns A and B. Column A has some missing values.
Write code to replace missing values in column A with the mean value of that column.

```python
##CREATING DATAFRAME WITH TWO COLUMNS

data=[['170',None],['165','18'],['172','16'],[None,'13'],['173',None],['156','17'],['155',None],['166','19'],['157','14'],['168','13']]
df=pd.DataFrame(data,columns=['Height','Age'])
```

```python
df
```

```python
df['Height'].isnull().sum()

```

```python
df['Height'].dtypes
```

```python
df['Height'] = df['Height'].fillna(0).astype(int)
```

```python
df['Height'].dtypes
```

```python
a=df['Height'].mean()
```

```python
a
```

```python
df['Height']=df['Height'].replace([0],a)
```

```python
df
```
